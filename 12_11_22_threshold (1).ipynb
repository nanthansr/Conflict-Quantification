```python
def Jaccard(str1, str2):
    str1 = set(str1.split())
    str2 = set(str2.split())
    return float(len(str1 & str2)) / len(str1 | str2)
```

```python
mls = []
for i in combine_and[0]:
  kls = []
  for j in combine_and[0]:
    try:
      kls.append(Jaccard(i[0],j[0]))
      try:
        kls.append(Jaccard(i[1],j[1]))
      finally:
        continue
        #print(i,j)
    except:
      continue
      #print(i,j)

    
      
  mls.append(kls)
```

```python
import random
random.sample(mls,1)
```

```python
len(mls[8300])
```

```python
matrix = np.matrix(mls)
```

```python
matrix.shape
```

```python
for i in range(len(matrix)):
  matrix[i][i] = 0
```

```python
for i in range(len(matrix)):
  for j in range(len(matrix)):
    if  (matrix.getA()[i][j]) > .999 :
        matrix.getA()[i][j] = 0
max_G = nx.DiGraph(np.matrix(matrix))
max_G.remove_nodes_from(list(nx.isolates(max_G)))
print(max_G.number_of_nodes())
```

```python
#threshold 0.001 - > 0.999
sat=[]
tval=[]
thresh=.001
while thresh<.999:
  for i in range(len(matrix)):
    for j in range(len(matrix)):
      if (matrix.getA()[i][j]) < thresh or (matrix.getA()[i][j]) > .999 :
          matrix.getA()[i][j] = 0
  G = nx.from_numpy_matrix(np.matrix(matrix))
  G.remove_nodes_from(list(nx.isolates(G)))
  sat.append(G.number_of_nodes())
  tval.append(thresh)
  thresh+=0.001
```

```python
xpoints = np.array(tval)
ypoints = np.array(sat)

plt.plot(xpoints, ypoints)
plt.show()
```

```python
#threshold 0.0001 - > 0.2

matrix = np.matrix(mls)
for i in range(len(matrix)):
  matrix.getA()[i][i] = 0
sat=[]
tval=[]
thresh=.00001
while thresh< .2 :
  for i in range(len(matrix)):
    for j in range(len(matrix)):
      if (matrix.getA()[i][j]) < thresh or (matrix.getA()[i][j]) > .999 :
          matrix.getA()[i][j] = 0
  G = nx.from_numpy_matrix(np.matrix(matrix))
  G.remove_nodes_from(list(nx.isolates(G)))
  sat.append(G.number_of_nodes())
  tval.append(thresh)
  thresh+=.00001
```

```python

xpoints = np.array(tval)
ypoints = np.array(sat)

plt.plot(xpoints, ypoints)
plt.show()
```
